## Clustering using some standard tools :::

Lets combine ratings, movies and tags data together and lets analyse more. This can be done using merging on 'movieId'

```python
movies_rating_df = pd.merge(movies_df, ratings_df, on='movieId')
movies_rating_df.head()
```

```python
mrt_df = pd.merge(movies_rating_df, tags_df, on='movieId')
mrt_df.head()
```

Since we are trying to recomend movie based on tags and rating lets drop other features out of the mrt dataFrame

```python
mrt_df = mrt_df[['movieId','rating', 'tag']]
mrt_df.head()
```

We have tag in the form of string variable, so lets appy some encoding

```python
le = preprocessing.LabelEncoder()
mrt_df = mrt_df.apply(le.fit_transform)
mrt_df.head()
```

Lets define a new method to plot a heatmap. So we can check corelation of movieId and other two features

```python
def plot_correlation(data):
    '''
    plot correlation's matrix to explore dependency between features 
    '''
    # init figure size
    rcParams['figure.figsize'] = 10, 10
    fig = plt.figure()
    sns.heatmap(data.corr(), annot=True, fmt=".2f")
    plt.show()
    fig.savefig('corr.png')
plot_correlation(mrt_df)
```

The result of the heatmap is suggesting that tag and rating are somehow more correlated. It means some movies of certain tags are getting really good rating value. Also we shall note the fact we havn't used any averaging of the colum i.e like the realtion between average rating value and tag is not plotted.

```python
kmeans = KMeans(n_clusters=10).fit(mrt_df)
centroids = kmeans.cluster_centers_
plt.scatter(mrt_df['movieId'], mrt_df['tag'], c= kmeans.labels_.astype(float), s=100, alpha=0.5)
plt.show()
```

Clearly from above K-Means scatter plot, we can see, we can divide data into different cluster using tags value. We can define tage as art with some conditions like containing word -'author', 'drama', 'book','romance' etc. After that we can pick any movie from art cluster and check its rating.

Now, I wanted to roughly project this tag scatter graph to the 3D projection thereby it would becomes more obvious how a movie can be selected. So I passed K-Means parameters themselves to the 3D GCA projection graph. Now we can see rating layers containing tag clusters. Which makes sense to pick a movie of anyrating and tag. 

```python
threedee = plt.figure().gca(projection='3d')
threedee.scatter(mrt_df.index, mrt_df['tag'], mrt_df['rating'], c= kmeans.labels_.astype(float), s=20, alpha=0.5)
threedee.set_xlabel('movies')
threedee.set_ylabel('Tag (Label Encoded)')
threedee.set_zlabel('rating (Label Encoded)')
plt.show()
```

So By using the means of clustering I showed how a movie can be suggested based on the rating and tags. We also searched more details of rating feature exploration which also usages the means of clustering to suggest a movie.
